 # *Best Fit line or linear regression line:
 
 The linear line which covers maximum data points(dots) in relationship b/w feature and label is called best fit line.
                                      
                                      OR
  Most of the data points close to any line , then that line is best fit line.


How do you know that this is best fit line?

The best best line is obtained by minimizing the residual. Residual is the distance b/w actual Y and the predicted y , as shown below:



Here:
blue dots are actual data points 

orange dots are predicted data points , predicted in such a way that distance b/w actual and predicted minimum (i.e residual).

and , line obtained is best fit line or linear regression line.


Mathematically , Residual is :

 
           r = y-(mx+c)   # i.e residual = actual-predicted -(residual for one point)
      
 Hence sum of square of residuals is:
    
          summation(ri)i=1-n = summation(yi-(mxi+c)       -(sum of residuals)
          
          R(x) = summation((ri)^2)= summation(yi-(mxi+c))^2  -(sum of squares of residuals)

# **Adjusted R2 Statistics Test: (Basically , works on multiple linear regression)

if we increase the number of independent variables(features , earlier we studied for only one) , then R2 value will also increases. But that doesn't mean that new independent variables will have any correlation with the output variable. 

In other words , even with the addition of new features in our model ,it is not necessary that our model yields better but yes R2 value will surely increase.

To solve this problem ,we use adjusted R2 value which pecifies excessive use of such features which do not co-relate with the input data .

let's understand this with example:


We can see that R2 always increases with an increase in the number of independent variables. Thus , it doesn't give a better picture and so we need  adjusted R2  value to keep this in check.

Mathematically , it is calculated as:(Behind the coding)



Here we can calculate R2 as studied earlier , and N is basically number of rows /Total sample size , as in above example there are 200 rows , so N=200. and p is number of predictors/features , as in above example we have 3 predictors / features i.e tv , radio, and newspaper.


In the equation above if p=0 , then adjusted R^2 will be equal to R^2. Thus adjusted R^2 , will always be less than or equal to R^2 ,and it penalises the excess of independent variables which do not affect the dependent variable.




# Why this is called adjusted R2 Test? and why we need this?

As number of features increases , R2 will also increases , so problem is , in just R2 statistic , model confidence cannot cross 100 % , it is just limited to 100 % only ,

But , if number of features increases , then there are chances that model confidence can cross 100% , but this is not correct , it must be less than or equal to 100% , then adjusted R2 test will punish and pull it down to 100% , that's why it is called adjusted.

  # Overfitting : 
  
  Your model is overfitting your training data when you see that the model performs well on the training data but does not perform well on the evaluation(test) data. This is because the model is memorizing the data it has seen and is unable to generalize to unseen examples.
  
Means , it will work perfectly on that data(or predict that data accurately) , which is properly understood by this(or seen that data) , which is wrong , if someone change the data then testing will not be so accurate.

So Here what we want,no matters  what kind of data we are giving (known or unknown) , it should give good accuracy in prediction. & model shuld not be overfit.

  
**What causes overfitting in models?**

Overfitting occurs when the model cannot generalize and fits too closely to the training dataset instead. Overfitting happens due to several reasons, such as: The training data size is too small and does not contain enough data samples to accurately represent all possible input data values.

**Let's see if our model is overfitting our training data using REGULARIZATION.**

# **REGULARIZATION:

When we use regression model to train our data , then there are good chances that the model will overfit the given training dataset.

**Regularization** helps to sort this overfitting problem by restricting the degrees of freedom of a given equation i.e simply reducing the number of degrees of polynomial function by reducing thier corresponding weights.

In a linear equation ,we don't want huge weught/coefficients , as small change in weight can make a large difference  for the dependent variable(Y). So regularization constraints the weight of such features  to avoid overfitting.

To regularize the model, a shrinkage penalty is added to the cost function.

Let's see different types of regulariztion in regression:


**LASSO**

**RIDGE**

**ELASTINCET(less popular)**





##

**1. LASSO-(Least absolute shrinkage and selection operator)-L1 Form**

LASSO regression penalizes the model based on the sum of magnitudes of the coefficients.The regularization term is given by:

regularization = λ* Σ |Bj| # (Here its beta)

where λ is an shrinkage operator.

**For example , we have three features in a dataset i.e phone number , exp. ,skill  and one label , i.e Salary**

Now , we need to predict the salary.

**So what LASSO does?**

It will start finding the pattern here , how exp., skill , phone number is related to salary.

So it found the relationship b/w skills and exp , as the skills ,exp. increases  , salary increases.

it is trying to find relationship b/w phone number and salary.

**if there is no relationship found b/w any feature and label , then LASSO will punish that entire column(feature) , as there is no relationship with label, so it will kick out that feature from dataset. and it will give zero importance to that feature which is of no use**





**2. RIDGE Regression (L2 Form):**

RIDGE regression penalizes the model based in the sum of square of mangintues of coefficients. The regularization term is given by :

Regularization =  λ * Σ |Bj^2|

where λ is an shrinkage operator.

**For example , If we consider the same example  , then unlike LASSO , RIDGE will give very less importance to phone number.**

It will keep that feature which shows no relationship with label , but will  give very less importance to that feature.

**3. Difference between LASSO & RIDGE regression:**

RIDGE regression shrinks the coefficients(percentage) for those features which contribute very less in ine model but they have huge weights  , very close to zero. But it never makes them exactly zero as LASSO does.

Thus the final model will still contain all the features , though with less weights. This doesn't help interpreting the model very well .

```python
from sklearn.linear_model import Ridge , Lasso , RidgeCV , LassoCV

# Here Ridge , Lasso are regression
# and RidgeCV , LassoCV ,will get learning rate (with how much speed our data should get trained)
```

**LASSO Regularization :**

```python
# Lasso will return best alpha(learning rate from LassoCV)  after max iteration
# Normalize is subtracting the  mean and divide by L2-form

lasscv = LassoCV(alphas = None , max_iter = 100 , normalize= True)
lasscv.fit(x_train,y_train)
```

max_iter means how much time data should learn the data , to get better prediction or to undestand the data better.

If you're giving a big number , then it will take more time to learn.

```python
# Best alpha parameter: Or learning rate ;

alpha = lasscv.alpha_
alpha
```

```python
# Now that we have best parameter ,Let's use Lasso regression & see how well our data has fitted before.

lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train , y_train)
```

```python
# Model learned everything and its time to test how good is test data?

lasso_reg.score(x_test,y_test)


# This model is good upto 79%
```
