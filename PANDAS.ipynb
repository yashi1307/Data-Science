## GroupBy:

GroupBy will group & we have to mention group by what column . 


```python
# variable = df.groupby(column,as_index= false/true)[which column information to be collect]

avg_sales = df.groupby('Company',as_index=False)['Sales'].mean()
avg_sales


# Here we got the mean of sales groupby company , it always gives in ascending order.

# as_index = false means want indexing , if  don't want just write as_index=True

# By default as_index = True , that means it does'nt show index.
```

```python
# Similarly we can find info of one or more column groupby any one column , and also find like mean ,sum etc.

total_sales = df.groupby('Company' , as_index= True)['Sales','Profit'].sum()
total_sales
```

```python
# if we want to one or more thing like mean , sum of one or more column groupby any one column.

sales = df.groupby('Company')['Sales','Profit'].agg(['mean','sum'])
sales
```

```python
## To find the Largest value in any column groupby any one column .

df.groupby('Company')['Sales'].nlargest(1)    # here 1 mean top 1 largest , & 2 with top 2 largest of each company.
```

```python
# To find the Smallest value in any column groupby any one column:

df.groupby('Company')['Sales'].nsmallest(1)
```

## Join: if you want to join by index

use this , when index is same but columns are different.

```python
left = pd.DataFrame({'A':['A0','A1','A2'],
                    'B':['B0','B1','B2']},
                  index= ['K0','K1','K2'])

right = pd.DataFrame({'C':['C0','C1','C2'],
                    'D':['D0','D1','D2']},
                  index= ['K0','K2','K3'])


# HERE index is also somewhat different in both DataFrames . 
```

```python
left
```

```python
right
```

```python
# DataFrame_1.join(DataFrame_2)

left.join(right)

# if we joining right(inside) to left , then left DataFrame data will not change that occurs complete ,

#but for righ,t only those index  which are same will occurs, for different it will give nan

# overall right dataframe has to come acc. to left , becoz left will complete.
```

```python
# DataFrame_2.join(DataFrame_1)

right.join(left)


# if we joining left(inside) to left , then right DataFrame data will not change that occurs complete ,

#but for left , only those index  which are same will occurs , for different it will give nan

# overall left dataframe has to come acc. to right , becoz right will come complete.
```

```python
# if we want everything , so can use 'outer':

left.join(right,how='outer')

# it will give everthing or all index . by default it is not outer it is how ='left'. chck by shift+tab , inside bracket.
```

## Apply Function:

```python
df = pd.DataFrame({'dept':[1,2,3,4,5],
                   'emp_name':['ravi','akash','aman','pri','prince'],
                   'salary':[1000,2000,3000,4000,5000]})
df
```

```python
df['new'] =df['salary'].apply(salary_hike)
print(df)



# here we added one more column to existing DataFrame.

# .apply function is used to apply some function , and we called salary_hike function.
```

```python
# if we want to find increased salary then:

df['increased_salary']= df['new']-df['salary']
df
```

```python
# variable_name = pd.date_range(starting_date(anything), periods(like how much montha , daya, years ), Frequency =M/D/Y(anything)

date_range = pd.date_range('13/7/2000' , periods = 12 , freq = 'M')

# CONVERT THIS TO DATAFREAME , IT WILL GIVE 12 MONTHS FROM THE GIVEN DATE:

# new_variable = pd.DataFrame(previous_variable , columns = ['any_column name(new column)'])
dates = pd.DataFrame(date_range, columns =['Date'])
dates


# it wil give data staring from month -7 , as we have given Frequency = M. & date will always be last date of month.
```

```python
# if we want data accoring to date , the have to set frequency = D

date_range = pd.date_range('13/7/2000' , periods = 12 , freq = 'D')

```

```python
dates = pd.DataFrame(date_range, columns =['Date'])
dates
```

```python
# If we want data according to year then have to set frequency = Y:
date_range = pd.date_range('13/7/2000' , periods = 12 , freq = 'Y')
```

```python
dates = pd.DataFrame(date_range, columns =['Date'])
dates
```

```python
# IF WE WANT TO SEPARATE DAY , MONTH & YEAR :

# variable_name[new-column name] = variable_name['column'].dt.corresponding value im column

dates['Day'] = dates['Date'].dt.day
dates['Month'] = dates['Date'].dt.month
dates['Year'] = dates['Date'].dt.year

dates
```

## nansum :
NumPy provides some special aggregations that will ignore missing values or NaN .

simply when you don't want to replace nan and want to add values too , then we should go for nansum.

it will not return nan as result.

```python
np.nansum(value1)  # sum of all values of value1 by avoiding nan values. ,it will not return nan as result.
```

                  1.isnull():(to check nan's are present or not)
                  2.notnull():(opposite of isnull): to check data is present or not
                  3.dropna(): drop or remove particular position with nan's or return the filtered version of data
                  4.fillna(): to fill data in place of nan's or return a copy of the data with missing values filled
                  
let us use one by one to understand these methods:

```python
# Another way:
 
df.isna().sum()   # na or null is same 
```

```python
# notnull() shows how many rows have not null (opposite of isnull())

df.notnull().sum()
```

## ## SQL DataFrame:



```python
import sqlite3

# connecting with the Database(if there no database exists , it will create one)

db = sqlite3.connect('my_database.db')

# declare cursor

cursor = db.cursor()
```

```python
# to create table : 

cursor.execute("CREATE TABLE stud_detail(phone_number INT, student_name TEXT , enrolled_date TEXT , marks INT)")
```

```python
# inserting records from CSV files:

with open("Student_details.csv",'r') as file:
    no_records=0 
    for row in file:
        cursor.execute("INSERT INTO stud_detail VALUES (?,?,?,?)",row.split(","))
        db.commit()
        no_records+=1
        
print(no_records,'Records_inserted')
            
```

```python
result = cursor.execute("SELECT * FROM stud_detail")

for row in result:
    print(row)
```
