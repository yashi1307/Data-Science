# Boosting

**Boosting is an ensemble approach(i.e it involves several trees) that starts from a weaker decisionand keeps on building the model such that the final prediction is the weighed sum of all the weaker decision makers . The weights are assigned based on the performance of individual Tree.**



Ensemble parameters are calculated in stagewise way which means that while calculating the subsequent weight , the learning from the previous tree is considered as well.


In short , In Bagging all models works parallely and the final prediction based on the majority of all models decision.

In Boosting , All models works one by one by passing prediction of previous to next and then final prediction is made.

**Weak classifier - why tree?**

First what is a weak classifier? 

**Weak classifier - slightly better than random guessing. Any algorithm could have been used as a base for the boosting technique, but the reason for choosing trees are:**

   **Pro's:**
     
    1.computational scalability,
    2.handles missing values, robust to outliers,
    3.does not require feature scaling i.e doesn'y need standarization tehniques.
    4.can deal with irrelevant inputs
    5.interpretable (if small)i.e if the data is small , then we can understand this as well.
    6.handles mixed predictors as well (quantitive and qualitative) i.e continous and categorical
    
    
   **Con's:**
    
    inability to extract a linear combination of features
    high variance leading to a small computational power
    
And that's where boosting comes into the picture. It minimises the variance by taking into consideration the results from various trees.
