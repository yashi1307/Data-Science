 # 3. **Working of Logistic Model:***

For linear Regression , the model is defined by: y = mx+c , 

and for logistic regression , we calculate probability i.e y is  the probability of given variable x .Thus it is obvious that value of y sholud lies in between 0 and 1.

But , when we use equation of linear regression to calculate probabilty , then  we will get values less than 0 and grater than 1. This doesm't makes any sense becoz we want categorical data(either 0 or 1).

So we need to use an such equation , whuch always gives  values between 0 and 1  , as we desire while calculating the probabilty.


**And for this internally it uses a function , Sigmoid Function(logic function)**

# 4. Sigmoid function:

**It will take that continuous number  , and will conver anythinng above 0.5 to 1 , and less than 0.5 to 0. so that we will get our desired data.**


**For example (same example ): As we use linear regression , then doctor will tell patient that he is 65% (0.65) infected , then sigmoid function will take that continuous data and will convert it into binary or categorical.**

As it is 0.65 , which is above than 0.5 , so it wil convert this 0.65 to 1 , and hence we can predict now that patient is infected to corona. , and if it comes less than 0.5 like 0.35 so it will convert to 0 , and we predicted that he is not infected to corona.

**Why do we use the Sigmoid Function?**

1. The sigmoid function range is bounded between 0 and 1 . Thus it is useful to find the probabilty of logistic Function.

2. It's derivative is easy to calculate than other functions which is useful during gradient descent function.

3. It is a simple way of introducing non-linearity to a model.

# # F-1 Score (Combination of Precision and Recall)

**F-1 Score is defined as the harmonic mean of Precision and Recall.**

The mathematical formula is :



**The F1-score combines the precision and recall of a classifier into a single metric by taking their harmonic mean. It is primarily used to compare the performance of two classifiers. **What does a high F1 score mean?**

The higher the precision and recall, the higher the F1-score. F1-score ranges between 0 and 1. The closer it is to 1, the better the model.

# ROC-(Reciever Operator Characterstics):

ROC curves in logistic regression are used for determining the best cutoff value for predicting whether a new observation is a "failure" (0) or a "success" (1).

**We know that the classification algorithms work on the concept of probability of occurrence of the possible outcomes. A probability value lies between 0 and 1.
zero means that there is no probability of occurrence and one means that the occurrence is certain.**

But while working with real-time data, it has been observed that we seldom(rarely) get a perfect 0 or 1 value. Instead of that, we get different decimal values lying between 0 and 1. 

Now the question is if we are not getting binary probability values(i.e 0 or 1).

how are we actually determining the class in our classification problem?

**There comes the concept of Threshold. A threshold is set, any probability value below the threshold is a negative outcome, and anything more than the threshold is a favourable or the positive outcome. For Example, if the threshold is 0.5, any probability  value below 0.5 means a negative or an unfavourable outcome and any value above 0.5 indicates a positive or favourable outcome.**

Now, the question is, what should be an ideal threshold?

The following diagram shows a typical logistic regression curve.




**The horizontal lines represent the various values of thresholds ranging from 0 to 1.**

Let's suppose our classification problem was to identify the obese(Fatty) people from the given data.

**The green markers represent obese people and the red markers represent the non-obese people.**

**Our confusion matrix will depend on the value of the threshold chosen by us.**

# # Check Multicolinearity problem , i.e find  if one feature is dependent on other feature

**Why to do this step or why to find relationship b/w features vs features?**

Let's understand this with an example: Let's consider we have a data of age , exp , and salary , here age and exp. are features , and salary is label . 


                Age              Exp.              Salary
                
                22                1                  4 LPA
                25                4                  9 LPA
                30                9                  15 LPA

Now , here it is clearly understood that when age increases , salary increases , which is positive relationship

when exp. increases , salary increase , which is also a positive relationship.

**But what's the relationship b/w age and exp.?**

It is clearly can be seen that , with in increase in age , exp. also increasing , and giving a positive relationship.

**So what if we only use one feature instead of two , because both of them are giving positive relationship wit label , and also  giving positive relationship b/w age and exxp.? So why should we are using extra features , if the trend is same with all ?**

So it'a better to delete extra feature , as extra is of no use , if one or all are giving same trend.


**Answer of question (why to use this step ) is : 1. If we are using extra features which is/are of no use(or no required) , then obviously these extra features takes more time to train , than to train only sufficient feature , because if training time increases , cost will increase to train a model.**

**2. if you're adding extra features (which are not required ) , then for sure model will be overfitted model .**

so to avoid these problem , we should check relationship b/w features and features (Multicolinearity problem) using **VIF(Variance Inflation Factor).**
